```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import matplotlib.pyplot as plt
```

```python
query = "arvidkahl"
limit = 5000
tweets = []
```

```python
tweets = []
for tweet in sntwitter.TwitterUserScraper(query).get_items():
		if len(tweets) == limit:
			break 
		else:
			tweets.append([tweet.date, tweet.id, tweet.content, tweet.url, tweet.likeCount, tweet.inReplyToTweetId])
			# tweets.append([vars(tweet)])
tweets
```

```python
tweet_df = pd.DataFrame(tweets, columns=['Date', 'Id', 'Tweet', 'Url', 'Likes', 'ReplyTo'])
tweet_df
```

```python
tweet_df = tweet_df[tweet_df['ReplyTo'].isnull()]
tweet_df
```

```python
df = tweet_df[['Likes']]
df
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
```

```python
plt.scatter(df.index, df['Likes'])
```

```python
X = np.array(df['Likes']).reshape(-1, 1)
X
```

```python
nbrs = NearestNeighbors(n_neighbors = 10)
nbrs.fit(X)
```

```python
# distances and indexes of k-neaighbors from model outputs
distances, indexes = nbrs.kneighbors(X)
# plot mean of k-distances of each observation
plt.plot(distances.mean(axis =1))
```

```python
outlier_index = np.where(distances.mean(axis = 1) > 10)
outlier_index
```

```python
outlier_values = df.iloc[outlier_index]
outlier_values
```

```python
# plot data
plt.scatter(df.index, df['Likes'], color = "b")
# plot outlier values
plt.scatter(outlier_values.index, outlier_values["Likes"], color = "r")
```
